```python
import sys
import os
import pandas as pd
from pandas import DataFrame

sys.path.insert(0, os.path.abspath('../src'))
```

```python
df_source = pd.read_csv('../raw/winemag-data_first150k.csv')
```

```python

def clean_columns(df:DataFrame)-> DataFrame:
    df = df_source.loc[df_source['country'] == 'Argentina']
    df = df.rename(columns={'designation':'Wine Name', 'description':'Description', 'country':'Country', 'points':'Points', 'price':'Price', 'province':'Province',   'variety':'Variety', 'winery':'Winery'})
    df = df.drop(columns=['Unnamed: 0'])
    df['region_full'] = df[['region_1', 'region_2']].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)
    df = df.drop(columns=['region_1', 'region_2'])
    return df
df = clean_columns(df_source)
```
